```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Prepare a range of x values for plotting the fitted curves
x_range = np.linspace(X.min(), X.max(), 1000).reshape(-1, 1)

# Create subplots for plotting
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.ravel()

for degree in range(1, 7):
    # Generate polynomial features
    poly = PolynomialFeatures(degree)
    X_train_poly = poly.fit_transform(X_train)
    X_range_poly = poly.transform(x_range)

    # Fit the Linear Regression model
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Predict outcomes
    y_range_pred = model.predict(X_range_poly)

    # Compute the training error
    y_train_pred = model.predict(X_train_poly)
    mse = mean_squared_error(y_train, y_train_pred)

    # Plot
    axs[degree - 1].scatter(X_train[:, 0], y_train, color='blue', s=10, label='Training points')
    axs[degree - 1].plot(x_range, y_range_pred, color='red', label=f'Degree {degree} fit (MSE: {mse:.2f})')
    axs[degree - 1].set_title(f'Polynomial Degree {degree}')
    axs[degree - 1].set_xlabel('X')
    axs[degree - 1].set_ylabel('y')
    axs[degree - 1].legend()

plt.tight_layout()
plt.show()
```

3. To increase the confidence of your answer, you must divide the data into training and test sets and make repeated runs with shuffled data (at least 20 runs). You must decide on the best way to make this decision. By using this approach, what is your decision and why? 

--- Your answer here --- 

```python
#--- Write your code here ---#
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline

# Create a list to store the average MSE for each degree
average_mse_per_degree = []

# Perform 20 runs with different shuffles for each polynomial degree
n_runs = 20
degrees = list(range(1, 7))

for degree in degrees:
    # Track the MSE for each shuffle
    mse_list = []
    
    for run in range(n_runs):
        # Create a shuffle split object
        shuffler = ShuffleSplit(n_splits=1, test_size=0.2, random_state=run)
        
        for train_idx, test_idx in shuffler.split(X):
            X_train, X_test = X[train_idx], X[test_idx]
            y_train, y_test = y[train_idx], y[test_idx]
            
            # Create a pipeline that generates polynomial features and fits a linear regression
            model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
            model.fit(X_train, y_train)
            
            # Evaluate the model on the test set
            y_pred = model.predict(X_test)
            mse = mean_squared_error(y_test, y_pred)
            mse_list.append(mse)
    
    # Compute the average MSE for this degree over all shuffles
    average_mse = np.mean(mse_list)
    average_mse_per_degree.append((degree, average_mse))

# Now select the degree with the lowest average MSE
best_degree, best_mse = min(average_mse_per_degree, key=lambda x: x[1])

print(f"The best model is of degree {best_degree} with an average MSE of {best_mse:.2f}")
```

```python
#--- Write your code here ---#

X_train, X_valid, y_train, y_valid = train_test_split(X_normalized, y, test_size=0.20, random_state=42)

# Define a simple logistic regression model
class SimpleLogisticRegression:
    def __init__(self, lr=0.01, iterations=1000):
        self.model = LogisticRegression(max_iter=iterations, solver='liblinear')

    def fit(self, X, y):
        self.model.fit(X, y)

    def predict(self, X):
        return self.model.predict(X)

# Initialize the forward selection process
model = SimpleLogisticRegression(lr=0.01, iterations=1000)
fs = ForwardSelection(X_train, y_train, model)

# Execute forward selection
fs.forward_selection()

print("Selected Features Indices:", fs.selected_features)

if any(idx >= X_valid.shape[1] for idx in fs.selected_features):
    print("Index out of range error: One or more selected feature indices exceed the column indices of X_valid.")


print("Shape of X_valid:", X_valid.shape)



try:
    y_pred = fs.predict(X_valid)
    roc_analysis = ROCAnalysis(y_pred, y_valid)
    f_score = roc_analysis.f_score()
    print("F-score on Validation Set:", f_score)
except Exception as e:
    print("An error occurred:", e)
```
